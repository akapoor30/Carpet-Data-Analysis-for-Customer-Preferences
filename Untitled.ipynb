```python
import pandas as pd
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import seaborn as sns

df = pd.read_csv('AM.csv', sep=',',encoding="mac_roman")

# Preprocess 
df['review_text'] = df['review_title'].astype(str) + ' ' + df['review_body'].astype(str)
df = df[['review_rating', 'review_text']]

aspects = ['quality', 'material', 'size', 'price']  
keywords = {
    'quality': ['quality','durable','good','decent'],
    'material': ['material','made','plastic','fabric'],
    'size': ['size','length','width','big'],
    'price': ['price','value','cost','expensive']
}

for key, words in keywords.items():
    df[key] = df['review_text'].apply(lambda x: any(word in x.lower() for word in words))
    
sid = SentimentIntensityAnalyzer()
df['sentiment'] = df['review_text'].apply(lambda x: sid.polarity_scores(x)['compound'])

# Visualization 1 
counts = df[aspects].sum().sort_values(ascending=False).reset_index()
counts.columns = ['aspect', 'count'] 
counts.plot.bar(x='aspect', y='count', rot=0, title='Reviews by Aspect of artificial mats')

# Visualization 2
sentiment_by_aspect = {}
for aspect in aspects:
    sentiment_by_aspect[aspect] = df[df[aspect]]['sentiment'].mean()

# Convert the sentiment by aspect into a DataFrame
sentiment_df = pd.DataFrame.from_dict(sentiment_by_aspect, orient='index', columns=['average_sentiment'])
sentiment_df.index.name = 'aspect'
sentiment_df.reset_index(inplace=True)

# Visualization 2: Bar plot of average sentiment score by aspect
plt.figure(figsize=(8, 6))
sns.barplot(data=sentiment_df, x='aspect', y='average_sentiment')
plt.title('Average Sentiment Score by Aspect of Artificial Mats')
plt.xlabel('Aspect')
plt.ylabel('Average Sentiment Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Creating a correlation matrix between aspects
correlation_matrix = df[aspects].corr()

# Visualizing the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
plt.title('Correlation between Aspects of Artificial Mats')
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```
