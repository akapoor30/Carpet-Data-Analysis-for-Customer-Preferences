COMPARISON BETWEEN THE ARTIFICIAL AND HANDMADE FLOOR MAT BY THEIR MATERIAL 

```python
import pandas as pd
from textblob import TextBlob
import nltk
import warnings
warnings.filterwarnings("ignore")
# import
# nltk.download('punkt')


aspects = ["material"]


def get_aspect_sentiments(df):
    aspect_sentiments = {aspect: [] for aspect in aspects}

    for review in df["review_body"]:
        # Convert review to string
        review = str(review)
        blob = TextBlob(review)

        for aspect in aspects:
            if aspect in review:
                sentence = next(
                    (
                        sentence
                        for sentence in blob.sentences
                        if aspect in str(sentence)
                    ),
                    None,
                )

                if sentence:
                   
                    polarity = sentence.sentiment.polarity

                    # Add the polarity to the aspect sentiments
                    aspect_sentiments[aspect].append(polarity)

    return aspect_sentiments


df1 = pd.read_csv("AM.csv", encoding="mac_roman")
df2 = pd.read_csv("HM.csv", encoding="mac_roman")

aspect_sentiments1 = get_aspect_sentiments(df1)
aspect_sentiments2 = get_aspect_sentiments(df2)


for aspect in aspects:
    print(
        f'Average sentiment for {aspect} in Artifical Floor mat: {sum(aspect_sentiments1[aspect]) / len(aspect_sentiments1[aspect])*100 if aspect_sentiments1[aspect] else "No reviews"}'
    )
    print(
        f'Average sentiment for {aspect} in Handmade floor mat: {sum(aspect_sentiments2[aspect]) / len(aspect_sentiments2[aspect])*100 if aspect_sentiments2[aspect] else "No reviews"}'
    )
```

```python
import matplotlib.pyplot as plt

# Calculate average sentiment for each aspect
avg_sentiment1 = (
    sum(aspect_sentiments1["material"]) / len(aspect_sentiments1["material"])
    if aspect_sentiments1["material"]
    else 0
)
avg_sentiment2 = (
    sum(aspect_sentiments2["material"]) / len(aspect_sentiments2["material"])
    if aspect_sentiments2["material"]
    else 0
)

# Prepare data for plotting
labels = ["Artificial Floor mat", "Handmade Floor mat"]
values = [avg_sentiment1, avg_sentiment2]

# Create a bar plot
plt.bar(labels, values, color=["blue", "green"])

# Add title and labels
plt.title("Average Sentiment for Material Aspect")
plt.xlabel("Floor Mat Type")
plt.ylabel("Average Sentiment")

# Show the plot
plt.show()
```

```python
import seaborn as sns

sns.violinplot(data=[aspect_sentiments1["material"], aspect_sentiments2["material"]])
plt.xticks([0, 1], ["Artificial Floor mat", "Handmade Floor mat"])
plt.ylabel("Sentiment Score")
plt.title("Violin Plot of Sentiment Scores for Material Aspect")
plt.show()
```

```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt

# Define aspects
aspects = [
    "material",
    "Quality",
    "brand",
    "price",
    "utility (propose of used)",
    "design",
    "made",
]


# Define a function to get aspect sentiments
def get_aspect_sentiments(df):
    aspect_sentiments = {aspect: [] for aspect in aspects}

    for review in df["review_body"]:
        # Convert review to string
        review = str(review)
        blob = TextBlob(review)

        for aspect in aspects:
            if aspect in review:
                # Get the sentence containing the aspect
                sentence = next(
                    (
                        sentence
                        for sentence in blob.sentences
                        if aspect in str(sentence)
                    ),
                    None,
                )

                if sentence:
                    # Get the polarity of the sentence
                    polarity = sentence.sentiment.polarity

                    # Add the polarity to the aspect sentiments
                    aspect_sentiments[aspect].append(polarity)

    return aspect_sentiments


# Get aspect sentiments for both datasets
df1 = pd.read_csv("AM.csv", encoding="mac_roman")
df2 = pd.read_csv("HM.csv", encoding="mac_roman")

aspect_sentiments1 = get_aspect_sentiments(df1)
aspect_sentiments2 = get_aspect_sentiments(df2)

# Initialize an empty dictionary to store average sentiments
avg_sentiments1 = {}
avg_sentiments2 = {}

# Calculate average sentiment for each aspect
for aspect in aspects:
    avg_sentiments1[aspect] = (
        sum(aspect_sentiments1[aspect]) / len(aspect_sentiments1[aspect])
        if aspect_sentiments1[aspect]
        else 0
    )
    avg_sentiments2[aspect] = (
        sum(aspect_sentiments2[aspect]) / len(aspect_sentiments2[aspect])
        if aspect_sentiments2[aspect]
        else 0
    )

# Prepare data for plotting
labels = aspects
values1 = [avg_sentiments1[aspect] * 100 for aspect in aspects]
values2 = [avg_sentiments2[aspect] * 100 for aspect in aspects]

# Create subplots
fig, axs = plt.subplots(2)

# Create a bar plot for Artificial Floor mat
axs[0].bar(labels, values1, color="blue")
axs[0].set_title("Artificial Floor mat")
axs[0].set_ylabel("Average Sentiment (%)")

# Create a bar plot for Handmade Floor mat
axs[1].bar(labels, values2, color="green")
axs[1].set_title("Handmade Floor mat")
axs[1].set_xlabel("Aspect")
axs[1].set_ylabel("Average Sentiment (%)")

# Rotate x-axis labels for better readability
for ax in axs:
    ax.set_xticklabels(labels, rotation=45, ha="right")

# Show the plots
plt.tight_layout()
plt.show()
```
